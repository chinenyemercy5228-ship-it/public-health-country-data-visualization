```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_countries = pd.read_csv("Group_work.csv")
print(df_countries)
```

```python
plt.figure(figsize=(8,5))

df_countries['category'].value_counts().plot(
    kind='bar',
    color=['skyblue', 'orange', 'green']
)

plt.title("Distribution of GDP Per Capita Categories")
plt.xlabel("Category")
plt.ylabel("Number of Countries")
plt.show()
```

```python
#no 2 Nigeria vs Algeria GDP over time
Nigeria = df_countries[df_countries['Country Name'] == 'Nigeria']
Algeria= df_countries[df_countries['Country Name'] == 'Algeria']

plt.figure(figsize=(8,5))

plt.plot(Nigeria['Year'], Nigeria['GDP'], marker='o', label='Nigeria')
plt.plot(Algeria['Year'], Algeria['GDP'], marker='o', label='Algeria')

plt.title("GDP Comparison: Nigeria vs Algeria")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.legend()
plt.show()
```

```python
#no 3 using a horizontal bar chart to represent the countries with highest GDP
latest_year = df_countries['Year'].max()

latest_data = df_countries[df_countries['Year'] == latest_year]

top10 = latest_data.nlargest(10, 'GDP')

plt.figure(figsize=(8,6))

plt.barh(top10['Country Name'], top10['GDP'])

plt.title(f"Top 10 Countries by GDP ({latest_year})")
plt.xlabel("GDP")
plt.ylabel("Country")


plt.show()
```

```python
#no 4 Scatter Plot of GDP Per Capita vs Education Expenditure

plt.figure(figsize=(8,5))

plt.scatter(
    df_countries['GDP Per Capita'],
    df_countries['Education Expenditure (% GDP)']
)

plt.title('GDP Per Capita vs Education Expenditure')
plt.xlabel('GDP Per Capita')
plt.ylabel('Education Expenditure (% GDP)')

plt.show()
```

```python
#no 5 Histogram showing the distribution of population density
plt.figure(figsize=(8,5))

plt.hist(
    df_countries['Population Density'],
    bins=20,
    edgecolor='black'
)
plt.text( 
4000, 
2000, 
"Positively Skewed", fontsize = 14)
plt.title("Distribution of Population Density")
plt.xlabel("Population Density")
plt.ylabel("Frequency")

plt.show()
```

```python
#6
plt.figure(figsize=(8,5))
import matplotlib.pyplot as plt

df_countries.boxplot(column = "GDP Per Capita", by = "Continent Name")

plt.title("GDP Per Capita Across Continents")
plt.xlabel("Continent")
plt.ylabel("GDP Per Capita")
plt.xticks(rotation=45)
plt.grid(False)

plt.show()
```

```python
#NO 7 Create a grouped bar chart showing the average Health Expenditure (% GDP) and 
#Education Expenditure (% GDP) for each continent.
continent_avg = df_countries.groupby('Continent Name')[
    ['Health Expenditure (% GDP)', 'Education Expenditure (% GDP)']
].mean()

continent_avg.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Average Health and Education Expenditure by Continent")
plt.ylabel("Percentage of GDP")

plt.show()
```

```python
#no 8 Plot a time series graph comparing the Population of Nigeria, Ghana, and South 
#Africa from the earliest year to the latest year in the dataset

countries = ['Nigeria','Ghana','South Africa']

for country in countries:
    temp = df_countries[df_countries['Country Name']==country]

    plt.plot(temp['Year'],
             temp['Population'],
             label=country)

plt.legend()

plt.title("Population Comparison")
plt.show()
```

```python
#no 9 
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

x = df_countries['Education Expenditure (% GDP)']
y = df_countries['Health Expenditure (% GDP)']

m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x+b, color='red')
plt.text(
    x.min(),
    y.max(),
    f'y = {m:.2f}x + {b:.2f}'
)
plt.title('Education vs Health Expenditure')
plt.xlabel('Education Expenditure (% GDP)')
plt.ylabel('Health Expenditure (% GDP)')

plt.show()
```

```python
#no 10
plt.figure(figsize=(8,5))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(df_countries['GDP Per Capita'], bins=10)
plt.title('Histogram of GDP Per Capita')
plt.text( 
50000, 
1500, 
"Positively Skewed", fontsize = 14)
# Box Plot
plt.subplot(1, 2, 2)
plt.boxplot(df_countries['GDP Per Capita'])
plt.title('Box Plot of GDP Per Capita')
plt.text( 
0.5, 
80000, 
"Outliers are extremely High", fontsize = 14)
plt.tight_layout()
plt.show()
```
